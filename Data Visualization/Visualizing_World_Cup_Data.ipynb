#Visualizing World Cup Data With Seaborn

For this project we will be exploring data from the Fifa World Cup from 1930-2014 to analyze trends and discover insights about the world’s game, fútbol!

This Fifa World Cup data is from [Kaggle](https://www.kaggle.com/abecklas/fifa-world-cup). Kaggle is a platform for data science competitions that hosts many datasets online.

Using Seaborn you will create a series of plots that explore aggregates and distribution across the goals scored in World Cup games.

##A little primer on the Fifa World Cup:

The FIFA World Cup, or simply the World Cup, is an international fútbol competition where 32 countries qualify to send teams made up of the best players from that nation to compete against each other for the World Cup championship.

The World Cup championship has been awarded every four years since the inaugural tournament in 1930, except in 1942 and 1946 when it was not held because of the Second World War.

The current format of the tournament involves 32 teams competing for the title at venues within the host nation over a period of one month.


```
#importing necessary libraries
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

```
#inspecting dataframe
df = pd.read_csv('WorldCupMatches.csv')
print(df.head())
```

```
df.columns
```

```
#Creating a new column in df named Total Goals, and set it equal to the sum of the columns Home Team Goals and Away Team Goals.
df['Total Goals'] = df['Home Team Goals'] + df['Away Team Goals']
print(df.head()) 
```

```
#creating a bar chart visualizing how many attendance were recorded each year the World Cup was held between 1930-2014.
sns.set_style('whitegrid')
sns.set_context('poster', font_scale=0.8)
f, ax=plt.subplots(figsize=(17,8))
sns.barplot(data=df, x = 'Year', y ='Attendance')
ax.set_title('Number of Attendance in world cup')
plt.show()
```

```
#creating a bar chart visualizing how many goals were scored each year the World Cup was held between 1930-2014.
sns.set_style('whitegrid')
sns.set_context('poster', font_scale=0.8)
f, ax=plt.subplots(figsize=(17,8))
ax = sns.barplot(data=df, x='Year', y='Total Goals')
ax.set_title('Total goals scored in each year')
plt.show()
```

```
#Loading goals.csv into a DataFrame called df_goals, and take a quick look at the DataFrame using .head().
df_goals = pd.read_csv('goals.csv')
print(df_goals.head())
```

```
#creating a box plot so we can visualize the distribution of the goals data instead of just the average with a bar chart.
sns.set_context('notebook', font_scale=1.25)
f, ax2 = plt.subplots(figsize=(17,7))
ax2 = sns.boxplot(data=df_goals, x='year', y='goals', palette='Spectral')
ax2.set_title('Box plot for goals scored in each year')
plt.show()
```
